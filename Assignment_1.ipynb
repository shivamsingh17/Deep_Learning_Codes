```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('Assignment1_Q4_data.txt')
```

```python
A = data[:500]
B = data[500:]
```

```python
plt.scatter(A[:, 0], A[:, 1], c = 'red', alpha = 0.5, label = 'class A')
plt.scatter(B[:, 0], B[:, 1], c = 'blue', alpha = 0.5, label = 'class B')
plt.title("Scatter plot of two classes")
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.show()
```
