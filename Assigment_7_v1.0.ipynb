```
def sigmoid(input):
    return 1.0 / (1.0 + np.exp(-(input)))
```

```
def error(Y, Y_hat): #Calculate the error
    err = 0.0
    #for x,y in zip(X,Y) :
        #fx = f(w, b, x)
    err += 0.5 * (Y_hat - Y) ** 2
    return err
```

```
e = []
for i in range(800):
  input = df.loc[i ,'Feature0': 'Feature39']
  m = input.values * w
  Z = np.sum(m)
  Y_hat = sigmoid(Z)
  e.append(error(Y[i], Y_hat))
  
print(e)
len(e)
```
