```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(-1, 1, 0.1)
```

```python
#Plotting h_11, h_12 and h_21 as three seprate figures.
Y11 = 1/(1 + np.exp(-400 * X - 24))
fig = plt.figure()
plt.title('H11 Plot')
plt.plot(X, Y11, linewidth = 3, color = 'r')
plt.xlabel('X')
plt.ylabel('Y11')
plt.grid()
plt.show()
```

```python
Y12 = 1/(1 + np.exp(-400 * X + 24))
fig = plt.figure()
plt.title('H12 Plot')
plt.plot(X, Y12, linewidth = 3, color = 'b')
plt.xlabel('X')
plt.ylabel('Y12')
plt.grid()
plt.show()
```

```python
Y21 = Y12 - Y11
fig = plt.figure()
plt.title('H21 Plot')
plt.plot(X, Y21, linewidth = 3, color = 'g')
plt.xlabel('X')
plt.ylabel('Y21')
plt.grid()
plt.show()
```
