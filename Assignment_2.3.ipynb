```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
X1 = np.arange(-5, 5, 0.1)
X2 = np.arange(-5, 5, 0.1)
X,Y = np.meshgrid(X1, X2)
Z11 = 1 / (1 + np.exp(-X - 100 * Y -200))
```

```python
#Sigmoid Function in 3D
fig = plt.figure()
plt.suptitle('Hill Plot')
ax = fig.add_subplot(1, 1, 1, projection = '3d')
ax.plot_surface(X, Y, Z11, rstride = 1,cstride = 1,color = 'r', antialiased = False)
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```
